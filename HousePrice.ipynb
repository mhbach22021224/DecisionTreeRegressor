```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('melb_data.csv')
df.head()
```

```python
df.columns
```

```python
#drop missing values
df = df.dropna(axis=0)
```

```python
#Selecting the prediction target
y = df.Price
#Select features
df_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = df[df_features]
```

```python
X.describe()
```

```python
X.info()
```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor

home_model = DecisionTreeRegressor(random_state=1)
home_model.fit(X,y)
```

```python
home_model.predict(X.head())
```

```python
y.head()
```

```python
from sklearn.metrics import mean_absolute_error
predicted_home_prices = home_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)

iowa_model = DecisionTreeRegressor(random_state=1)
iowa_model.fit(train_X, train_y)

val_predictions = iowa_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

```python
y.mean()
```

Optimizing

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
  model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=1)
  model.fit(train_X, train_y)
  val_pred = model.predict(val_X)
  mae = mean_absolute_error(val_y, val_pred)
  return mae
```

```python
for max_leaf_nodes in [5, 50, 100, 250, 500, 5000]:
  my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
  print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
scores = {leaf_size: get_mae(leaf_size, train_X, val_X, train_y, val_y) for leaf_size in [5, 50, 100, 250, 500,5000]}
best_tree_size = min(scores, key=scores.get)
print(best_tree_size)
```

```python
#Last model
final_model = DecisionTreeRegressor(max_leaf_nodes=best_tree_size, random_state=1)
final_model.fit(X, y)

mean_absolute_error(y, final_model.predict(X))
```
